```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mlp 
import seaborn as sns
```

```python
df = pd.read_csv('marvel.csv') 
df
```

Aqui podemos ver todas las variables que tenemos al igual que el tamaño de nuestra base de datos que es de 610 renglones x 10 columnas. Tenemos variables que describen a los personajes como su inteligencia, su fuerza, velocidad, durabilidad, poder y combate, igualmente tenemos su nombre y si son buenos, malos o neutrales. La variable que estaremos analizando en este codigo sera la de Strength ya que considero que el atributo mas importante de cualquier villano o superheroe es la Fuerza.

```python
df.describe()
```

Al utilizar df.describe(), podemos entender un poco mas nuestras variables ya que nos muestra varios conceptos como count, mean, minimo, maximo y el promedio. Podemos ver que todas las variables tienen un minimo de 0 o 1 un maximo de 100 a 120, esto nos muestra que en la base de datos tenemos todo tipo de heroes y villanos que puede que sean muy buena para alguna cosa pero muy mala para otra. Podemos ver que la variable Strength tiene una media de 29.03, un minimo de 1 y un maximo de 100.

```python
df.info()
```

Aqui podemos ver que tenemos 8 variables numericas indicadas por int64 y 2 variables categoricas indicadas por object. Esto nos sirve ya que podemos ver que variables podemos utilizar para que cosas ya sea graficas, correlaciones, etc. 

```python
df.corr()
```

Utilizando df.corr() podemos ver la correlación entre todas las variables. Esto nos sirve para ver que variables tienen mucha conexión con otras variables. Podemos ver que la variable que mas correlación tiene con Fuerza es Durabilidad, tienen una correlación de 0.793408. 

```python
sns.scatterplot(x='Strength',
                y='Durability',
                data=df)
plt.title('Correlación entre Durabilidad y Fuerza', fontsize = 18) 
plt.show()
```

En esta grafica podemos ver que existe una correlación entre Durabilidad y Fuerza de los heroes o villanos de Marvel. Entre mayor Durabilidad tenga un personaje, es muy probable que se Fuerza suba y viceversa. Con esto podemos hacer nuestra función de regresión la cual nos ayudara a pronosticar la Fuerza de cualquier personaje con la Durabilidad. La función seria Strength = a + b(Durability).

```python
df['Strength'].plot(kind='hist', 
 figsize=(10,10))
```

Gracias a este histograma de la variable Strength podemos ver que la mayoria de nuestros heroes y villanos tienen una Fuerza muy baja de entre 0 y 10. Podemos observar que no es un histograma normal ya que no tiene la forma de una campana.

```python
import statsmodels.api as sm
```

```python
model = sm.OLS.from_formula('Strength ~ Durability',
                            data = df) 
result = model.fit()

result.summary()
```

Aqui podemos ver una regresión con una ajuste de R igual a aproximadamente 63%. Un valor deseado en este aspecto es entre 60%-90%. Aunque entramos dentro de esta categoria, sigue siendo muy bajo y se gustaria encontrar un mejor R cuadrado para que sean mas exacto los resultados. Igualmente nos da los valores necesarios para completar la función para pronosticar la fuerza de un personaje basado en la durabilidad, esta seria Strength= -0.5628 + 0.7090(Durabilidad). 

# Pregunta 1 ¿Como es el promedio de Fuerza dependiendo de la alineación(bueno, malo o neutral)?

```python
pd.pivot_table(df,index='Alignment'
               , values='Strength',
              aggfunc = ['mean'])
```

Podemos ver que los personajes neutrales tienen el promedio mas alto de Fuerza con un 45.73, despuessiguen los malos con un promedio de Fuerza de 34.13 y  finalmente los buenos con un promedio de 26.64.

# Pregunta 2 ¿Que alineación tiene la mayor inteligencia?

```python
plt.figure(figsize=(4,4)) 
sns.scatterplot(x='Intelligence', y='Alignment', data=df)
plt.title('Inteligenica por Alineación', fontsize = 20) 
plt.show
```

En esta grafica de dispersión podemos obesevrar que la alineación con mejor inteligencia es la de los malos. Tambien podemos ver que los neutrales tienen un buen promedio de inteligencia ya que no tienen a nadie con una inteligencia menor a aproximadamente 38.

# Pregunta 3 ¿La alineación de los personajes afecta su habilidad de combate?

```python
plt.figure(figsize=(10,10)) 
sns.catplot(x='Alignment', y='Combat',kind='bar', data=df, ci = None);
plt.show
```

Como podemos ver en esta grafica de barras, la alineación si afecta en la habilidad de combate de los personajes neutrales, sin embargo, podemos ver que hay una diferenia insignificante entre el combate de los buenos y los malos, entonces en ese caso, la alineación no afecta su habilidad de combate.

# Pregunta 5 ¿Como es el promedio, el maximo y el minimo de Combate y Poder de cada alineación al igual que de toda la base de datos?

```python
my_pivot = pd.pivot_table(df,
                          index = 'Alignment',
                          values = ['Combat', 'Power'],
                          aggfunc = ['min', 'max', 'mean'],
                          margins = True,
                         )
my_pivot
```

Analizando esta Tabla Dinamica podemos ver que el minimo de los malos, buenos y el total de Poder y Combate es de 0, el minimo de los neutrales es de 12. El maximo de Poder y de Combate de los malos, los buenos, los neutrales y de todos es de 100. El promedio de Combate de los malos es de 44.25, de los buenos es 42.22, de los neutrales es de 67.64 y de todos los personajes es de 43.24. El promedio de Poder de los malos es de 44.05, de los buenos es de 38.26, de los neutrales es de 58.82 y de todos los personajes es de 40.21.
