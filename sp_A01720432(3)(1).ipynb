```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mlp 
import seaborn as sns
```

```python
df = pd.read_csv('marvel.csv') 
df
```

Aqui podemos ver todas las variables que tenemos al igual que el tamaño de nuestra base de datos que es de 610 renglones x 10 columnas. Tenemos variables que describen a los personajes como su inteligencia, su fuerza, velocidad, durabilidad, poder y combate, igualmente tenemos su nombre y si son buenos o malos. La variable que estaremos analizando en este codigo sera la de Strength ya que considero que el atributo mas importante de cualquier villano o superheroe es la Fuerza.

```python
df.describe()
```

Al utilizar df.describe(), podemos entender un poco mas nuestras variables ya que nos muestra varios conceptos como count, mean, minimo, maximo y el promedio. Podemos ver que todas las variables tienen un minimo de 0 o 1 un maximo de 100 a 120, esto nos muestra que en la base de datos tenemos todo tipo de heroes y villanos que puede que sean muy buena para alguna cosa pero muy mala para otra. Podemos ver que la variable Strength tiene una media de 29.03, un minimo de 1 y un maximo de 100.

```python
df.info()
```

Aqui podemos ver que tenemos 8 variables numericas indicadas por int64 y 2 variables categoricas indicadas por object. Esto nos sirve ya que podemos ver que variables podemos utilizar para que cosas ya sea graficas, correlaciones, etc. 

```python
df.corr()
```

Utilizando df.corr() podemos ver la correlación entre todas las variables. Esto nos sirve para ver que variables tienen mucha conexión con otras variables. Podemos ver que la variable que mas correlación tiene con Fuerza es Durabilidad, tienen una correlación de 0.793408. 

```python
sns.set_context('notebook')
sns.scatterplot(x='Durability',
                y='Strength',
                data=df)
plt.title('Correlación entre Durabilidad y Fuerza') 
plt.show()
```

En esta grafica podemos ver que existe una correlación entre Durabilidad y Fuerza de los heroes o villanos de Marvel. Entre mayor Durabilidad tenga un personaje, es muy probable que se Fuerza suba y viceversa. Con esto podemos hacer nuestra función de regresión la cual nos ayudara a pronosticar la Fuerza de cualquier personaje con la Durabilidad. La función seria Strength = a + b(Durability).

```python
df['Strength'].plot(kind='hist', 
 figsize=(10,10))
```

Gracias a este histograma de la variable Strength podemos ver que la mayoria de nuestros heroes y villanos tienen una Fuerza muy baja de entre 0 y 10. Podemos observar que no es un histograma normal ya que no tiene la forma usual de una campana.

```python
import statsmodels.api as sm
```

```python
model = sm.OLS.from_formula('Strength ~ Durability',
                            data = df) 
result = model.fit()
result.summary()
```
